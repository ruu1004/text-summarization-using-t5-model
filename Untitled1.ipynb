```python
# Load the dataset
dataset = load_dataset("csv", data_files="/content/test.csv")
# Model and tokenizer setup
model_name = "t5-base"
model = T5ForConditionalGeneration.from_pretrained(model_name)
tokenizer = T5Tokenizer.from_pretrained(model_name)
```

```python
def preprocess_function(examples):
    # Input text (e.g., article)
    inputs = examples['article']  # Ensure 'article' is the correct column name in your dataset
    # Tokenize the inputs
    model_inputs = tokenizer(inputs, max_length=512, truncation=True, padding="max_length")
    # Tokenize the labels (e.g., highlights)
    labels = tokenizer(
        examples['highlights'],  # Ensure 'highlights' is the correct column name
        max_length=128,
        truncation=True,
        padding="max_length"
    )
    # Add the labels to the model inputs
    model_inputs["labels"] = labels["input_ids"]
    return model_inputs
# Apply preprocessing to the dataset
encoded_dataset = dataset.map(preprocess_function, batched=True)

```

```python
print(encoded_dataset)
split_dataset = encoded_dataset["train"].train_test_split(test_size=0.1, seed=42)
train_dataset = split_dataset["train"]
validation_dataset = split_dataset["test"]
```

```python
# Assume a 90-10 split for training and validation
train_size = int(0.9 * len(encoded_dataset["train"]))
train_dataset = encoded_dataset["train"].shuffle(seed=42).select(range(train_size))
validation_dataset = encoded_dataset["train"].shuffle(seed=42).select(range(train_size, len(encoded_dataset["train"])))

print(f"Train Dataset Size: {len(train_dataset)}")
print(f"Validation Dataset Size: {len(validation_dataset)}")
```

```python
from transformers import T5ForConditionalGeneration, T5Tokenizer
from torchviz import make_dot
# Load T5 model and tokenizer
model = T5ForConditionalGeneration.from_pretrained("t5-small")
tokenizer = T5Tokenizer.from_pretrained("t5-small")
# Input text for the encoder
input_text = "translate English to French: Hello, how are you?"
inputs = tokenizer(input_text, return_tensors="pt")
# Dummy target text for the decoder (needed for logits calculation)
decoder_input_text = "Bonjour, comment allez-vous?"  # Target sequence prefix
decoder_inputs = tokenizer(decoder_input_text, return_tensors="pt")
# Add `decoder_input_ids` to the inputs
inputs.update(decoder_input_ids=decoder_inputs["input_ids"])
# Forward pass with both encoder and decoder inputs
outputs = model(**inputs)
# Visualize the computation graph of the logits
logits = outputs.logits
dot = make_dot(logits, params=dict(model.named_parameters()))
dot.render("t5_model", format="png", cleanup=True)
```

```python
from transformers import T5ForConditionalGeneration, T5Tokenizer
from rouge_score import rouge_scorer
import matplotlib.pyplot as plt
# Load the T5 model and tokenizer
model = T5ForConditionalGeneration.from_pretrained("t5-small")
tokenizer = T5Tokenizer.from_pretrained("t5-small")
# Define input texts and references
inputs = [
    "summarize: The quick brown fox jumps over the lazy dog. This is an example sentence.",
    "summarize: Transformers models are very powerful for natural language processing tasks."
]
references = [
    "A quick brown fox jumps over a dog.",
    "Transformers are powerful for NLP tasks."
]
# Generate predictions
generated_texts = []
for text in inputs:
    input_ids = tokenizer(text, return_tensors="pt").input_ids
    outputs = model.generate(input_ids)
    generated_texts.append(tokenizer.decode(outputs[0], skip_special_tokens=True))
# Calculate ROUGE scores
rouge_1_scores, rouge_2_scores, rouge_3_scores,rouge_l_scores = [], [], [], []
scorer = rouge_scorer.RougeScorer(["rouge1", "rouge2", "rouge3", "rougeL"], use_stemmer=True)
for ref, gen in zip(references, generated_texts):
    scores = scorer.score(ref, gen)
    rouge_1_scores.append(scores["rouge1"].fmeasure)
    rouge_2_scores.append(scores["rouge2"].fmeasure)
    rouge_3_scores.append(scores["rouge3"].fmeasure)
    rouge_l_scores.append(scores["rougeL"].fmeasure)
# Print ROUGE scores
for i, (r1, r2, r3, rl) in enumerate(zip(rouge_1_scores, rouge_2_scores, rouge_3_scores, rouge_l_scores)):
    print(f"Sample {i+1}:")
    print(f"  ROUGE-1: {r1:.4f}")
    print(f"  ROUGE-2: {r2:.4f}")
    print(f"  ROUGE-L: {rl:.4f}")
    print(f"  rouge-3: {r3:.4f}")
```

```python
# Plot ROUGE scores
x = [f"Sample {i+1}" for i in range(len(inputs))]
plt.figure(figsize=(10, 6))
plt.plot(x, rouge_1_scores, label="ROUGE-1", marker="o")
plt.plot(x, rouge_2_scores, label="ROUGE-2", marker="o")
plt.plot(x, rouge_3_scores, label="ROUGE-3", marker="o")
plt.plot(x, rouge_l_scores, label="ROUGE-L", marker="o")
plt.xlabel("Samples")
plt.ylabel("ROUGE Score")
plt.title("ROUGE Scores for T5 Model Outputs")
plt.legend()
plt.grid()
plt.show()
```
